```python
# Whch exchanges are supported
import ccxt
exchanges = ccxt.exchanges

print(exchanges)
```

```python
# of in een nette lijst
for exchange in exchanges:
    print(exchange)
```

```python
# ccxt in asynchronous mode
# import ccxt.async_support as ccxt # link against the asynchronous version of ccxt
```

```python
# Bitvavo gegevens bekijken (nog geen api key nodig)
bybit   = ccxt.bybit()

bybit_markets = bybit.load_markets()

# print(bybit.id, bybit_markets)
print(bybit_markets)
```

In Asynchronous mode krijg je bovenstaande melding, dus maar niet doen

```python
for bybit_market in bybit_markets:
    print(bybit_market)
```

```python
# Print only EURO pairs
for bybit_market in bybit_markets:
    if bybit_market.endswith('USDT'):
        print(bybit_market)
    
```

```python
# bekijken van order book
print(bybit.fetch_order_book(bybit.symbols[0]))
```

```python
# Bekijken ticker van BTC
print(bybit.fetch_ticker('BTC/USDT'))
```

```python
# Bekijken van trades
print(bybit.fetch_trades('LTC/USDT'))
```
