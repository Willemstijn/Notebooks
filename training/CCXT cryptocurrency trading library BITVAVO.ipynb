````python
# Whch exchanges are supported
import ccxt
exchanges = ccxt.exchanges

print(exchanges)
````

````python
# Bitvavo gegevens bekijken (nog geen api key nodig)
bitvavo   = ccxt.bitvavo()

bitvavo_markets = bitvavo.load_markets()

print(bitvavo.id, bitvavo_markets)
````

````python
hitbtc   = ccxt.hitbtc({'verbose': True})
hitbtc_markets = hitbtc.load_markets()
print(hitbtc.id, hitbtc_markets)
````

### Downloading data from bitvavo with freqtrade

This exchange supports: ['1m', '5m', '15m', '30m', '1h', '2h', '4h', '6h', '8h', '12h', '1d']

``freqtrade download-data --config user_data/configs/config_bitvavo.json --days 9 -t 1m 5m 15m 30m 1h 2h 4h 6h 8h 12h 1d``

of

``freqtrade download-data --config user_data/configs/config_bitvavo.json --days 1200 -t 30m 1h 2h 4h 6h 8h 12h 1d``
