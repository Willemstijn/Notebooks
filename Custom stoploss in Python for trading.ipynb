Hieronder wat voorbeelden die ik nageprogrammeerd van een of andere hindi trading channel over custom stoplosses.

```python
# Importeer de benodigde bibliotheken
import pandas as pd
import pandas_ta as pta
from pandas import DataFrame
import numpy as np

# Load data (LET OP: Ik gebruik hier 30min candles maar dit worden straks 5 min candles)
df = pd.read_json("./data/BTC_USDT-30m.json")
df.columns=['date','open','high','low','close','volume']
df['date']=(pd.to_datetime(df['date'],unit='ms'))
df.tail(5)
```

```python
# Calculate a stoploss level based on the first dataframe row close

# Load the dataframe column df['close'] as a list in variable 'close'
close = df['close']
# Take the first entry of the list to calculate the Stoploss
entry = close[0]
initialSL = entry + 60
trail = 10
forwardmove = 20
SL = initialSL
print('Entry: ', entry, 'SL: ', initialSL)
```

```python
df.head(5)
```

```python
for i in range(0, len(close)-1):
    diff = round((entry - close[1]),2)
    point = int(diff/forwardMove) if diff > 0 else 0
    SLmove = point*trail
    
    if initialSL - SL < SLmove:
        SL = initialSL - SLmove # SL order update here
        print(f'{i } {df.loc[i]["date"]} close {close[i]} New SL {SL} diff {diff}') # {point} {SLmove} ')'
    elif SL <= close[i]:
        print('SL hit: ', df.loc[i]['date'], close[i], '', SL)
        break
    else:
        print(f'{i } {df.loc[i]["date"]} close {close[i]} SL {SL} diff {diff}')
```
