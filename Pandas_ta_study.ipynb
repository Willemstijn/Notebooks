## Kolommen toevoegen voor bepalen Long / Short positie

```python
import numpy as np

# SMA check
df['SMA_pos'] = np.where(df['close']>df['SMA_23'], 'long' , 'short')
# MFI check
df['MFI_pos'] = np.where(df['MFI_14']>50, 'long', 'short')
# RSI check
df['RSI_pos'] = np.where(df['RSI_14']>55, 'long', 'short')
```

```python
df[['# openTime','close','SMA_23','SMA_pos','MFI_14','MFI_pos','RSI_14','RSI_pos']].tail(10)
```

## Add last column for checking if all conditions are 'long'

Eerst moeten de long en short condities omgezet worden in cijfers zodat deze opgeteld kunnen worden. 

0 = short  
1 = long  

Bij positie waarde 0 = short  
Bij positie waarde 1 = neutral  
Bij positie waarde 2 = warning  
Bij positie waarde 3 = long  

```python
# SMA check
df['SMA_pos'] = np.where(df['close']>df['SMA_23'], 1 , 0)
# MFI check
df['MFI_pos'] = np.where(df['MFI_14']>50, 1 , 0)
# RSI check
df['RSI_pos'] = np.where(df['RSI_14']>55, 1, 0)

df[['# openTime','close','SMA_23','SMA_pos','MFI_14','MFI_pos','RSI_14','RSI_pos']].tail(10)
```

```python
df['position'] = df['SMA_pos'] + df['MFI_pos'] + df['RSI_pos']

df[['SMA_pos','MFI_pos','RSI_pos','position']].tail(10)
```
